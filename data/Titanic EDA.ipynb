```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from collections import namedtuple

dataset = namedtuple('data', 'trainset, validation, test') 
```

```python
trainset = pd.read_csv('train.csv', low_memory=False)
testset = pd.read_csv('test.csv', low_memory=False)
gendermodel = pd.read_csv('gendermodel.csv', low_memory=False)
genderclassmodel = pd.read_csv('genderclassmodel.csv', low_memory=False)

data = dataset._make([trainset[:int(len(trainset.index)*0.8)], 
               trainset[int(len(trainset.index)*0.8):],
               testset])
```

```python
data.trainset.Sex = data.trainset.Sex.apply(lambda x: 1 if x == 'female' else 0)
data.validation.Sex = data.trainset.Sex.apply(lambda x: 1 if x == 'female' else 0)
```

```python
data.trainset.Embarked.unique()
```

```python
replacer = {'S': 10, 'C': 11, 'Q': 12}
data.trainset.Embarked = data.trainset.Embarked.apply(lambda x: replacer[x] if pd.notnull(x) else np.nan)
data.validation.Embarked = data.validation.Embarked.apply(lambda x: replacer[x] if pd.notnull(x) else np.nan)
```
